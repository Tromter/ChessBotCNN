```python
#here we are going to load our full chess data set.
#this looks in our directory and finds the processed_data_test.npz that has already
#been processed for this demo
chess_dataset = Chess_Data_Load()

#if you would like to look at the data you can do so here:
#chess_dataset.show_x_y()

#we get our xtr and ytr sets
data_x = chess_dataset.get_x()
data_y = chess_dataset.get_y()

#we load our model
model = tensorChess()

#we are using the Adam optimization
#the reasoning behind choosing Adam that its not only easy but its computationally cheap
#because this process is being tested on my little laptop I chose to use the fastest 
#optim
optim = tf.keras.optimizers.Adam()

#compile the model we have built
#we are using a "mean_squared_error" loss which means the closer to 0 our training set is
#the better
model.compile(loss='mean_squared_error', batch_size=256, optimizer = optim)


#train your model!
model.fit(data_x, data_y)

#print the model summary so you can see the layers at work
model.summary()
```

```python
#we will save this model so it can be used by move_eval class
model.save("save_model/my_model")
```

```python
#lets play a game!
#here we have a simple leaf exploration function, where we take the list of all
#available moves from a current state, and check each of their "values" using the trained
#CNN.  This list of tuples (value, move) is then returned
def explore_leaves(s, v):
    x = []
    for e in s.edges():
        s.board.push(e)
        x.append((v(s), e))
        s.board.pop()
    return x

#a getkey function to just get the value for sorting
def getKey(item):
    return item[0]
    
```
